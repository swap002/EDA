```
# Connect to the database.
connection = pymysql.connect(host='projects.insaid.co',
                             user='student',
                             password='student',                             
                             db='Capstone2',
                             charset='utf8mb4',
                             cursorclass=pymysql.cursors.DictCursor)
 
print ("connect successful!!")

try:
  
 
    with connection.cursor() as cursor:
       
        # SQL 
        sql = "SELECT count(dept_id) FROM department_data "
         
        # Execute query.
        cursor.execute(sql)
         
         
        for row in cursor:
            print(row)
             
finally:
    # Close connection.
    connection.close()
```
